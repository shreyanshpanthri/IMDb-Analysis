# Data Retrieval in Pandas

Data retrieval refers to accessing or selecting specific data from a DataFrame or Series for analysis and manipulation.

## Ways of Data Retrieval

### 1. Indexing

* Used to access a single column or element.
* Retrieves data using labels or column names.
* Suitable for selecting one column at a time.

### 2. Slicing

* Used to retrieve a range of rows or columns.
* Allows selection of a subset of data.
* Commonly uses start and end positions.

### 3. Fancy Indexing

* Used to retrieve multiple specific rows or columns simultaneously.
* Selection is performed using a list of labels or positions.
* Useful when required data is not in a continuous range.

---

# Functions Used for Data Retrieval

## `head()`

* Displays the first few rows of the dataset.
* By default, returns the first 5 rows.
* Useful for quickly inspecting the dataset structure.

## `tail()`

* Displays the last few rows of the dataset.
* By default, returns the last 5 rows.
* Helps examine the ending records of the dataset.

## `sample()`

* Returns randomly selected rows from the dataset.
* Useful for creating random samples and quick data inspection.

## `loc[]`

* Label-based indexing method.
* Retrieves rows and columns using their names or labels.
* Includes both start and end labels while slicing.

## `iloc[]`

* Integer position-based indexing method.
* Retrieves rows and columns using numerical positions.
* Excludes the ending position while slicing, similar to Python indexing.

---

## Quick Summary

| Method         | Purpose                               |
| -------------- | ------------------------------------- |
| Indexing       | Access specific columns or elements   |
| Slicing        | Retrieve a continuous range of data   |
| Fancy Indexing | Select multiple specific rows/columns |

### Single Column Retrieval

**Dot Notation (`.`)**

* Access a column using the column name.
* Syntax:


**Square Bracket Notation (`[]`)**

* Access a column using its name inside brackets.
* Syntax:


**Note:** Both return a **Series**. Square bracket notation is more flexible and preferred.

## Condition-Based Filtering

* Used to retrieve rows that satisfy a specific condition.
* Returns only the records matching the condition.

# String Operations in Pandas

String operations are used to manipulate and analyze text data stored in DataFrame columns. These operations can only be applied to columns with the **object (string)** data type.

## Why String Operations?

Many real-world datasets contain textual information such as movie titles, genres, actor names, and descriptions. String functions help clean, standardize, and transform this data for analysis.

## Data Type Conversion

Before applying string methods, ensure the column is of string type.

```python
df['column_name'] = df['column_name'].astype(str)
```

Example:

```python
df['runtime'] = df['runtime'].astype(str)
```

## Common String Methods

### Convert Text to Uppercase

```python
df['title'].str.upper()
```

Converts all characters to uppercase.

### Convert Text to Lowercase

```python
df['title'].str.lower()
```

Converts all characters to lowercase.

### Swap Case

```python
df['title'].str.swapcase()
```

Changes uppercase letters to lowercase and vice versa.

### Title Case

```python
df['title'].str.title()
```

Capitalizes the first letter of each word.

### Capitalize

```python
df['title'].str.capitalize()
```

Capitalizes only the first character of the string.

### Check Alphanumeric Values

```python
df['title'].str.isalnum()
```

Returns True if all characters are alphanumeric.

## Applications in IMDb Dataset

String operations can be used to:

* Standardize movie titles
* Clean genre information
* Format actor and director names
* Remove unwanted characters
* Prepare text data for analysis and visualization

These operations are essential during the data cleaning and preprocessing stage of Exploratory Data Analysis (EDA).

# Runtime Analysis

## Overview

The runtime column was converted from object datatype to numeric datatype using `pd.to_numeric()` with `errors='coerce'` to handle invalid values. This enabled statistical analysis and visualization of movie durations.

```python
df['runtime'] = pd.to_numeric(df['runtime'], errors='coerce')
```

## Descriptive Statistics

| Statistic          |      Value |
| ------------------ | ---------: |
| Count              |       1421 |
| Mean               | 133.24 min |
| Median             |    133 min |
| Standard Deviation |  24.78 min |
| Minimum            |     17 min |
| 25th Percentile    |    118 min |
| 75th Percentile    |    148 min |
| Maximum            |    321 min |

## Distribution Analysis

The Kernel Density Estimation (KDE) plot shows that movie runtimes follow an approximately bell-shaped distribution centered around 130–140 minutes.

Key observations:

* Most movies have runtimes between 110 and 150 minutes.
* The peak density occurs around 130–140 minutes, indicating that this is the most common runtime range.
* The distribution exhibits a slight positive skew, suggesting the presence of longer-duration movies.
* Extremely short and extremely long movies are relatively uncommon.

## Skewness Analysis

**Skewness = 0.956**

Interpretation:

* A skewness value close to +1 indicates a moderately right-skewed distribution.
* Most movies are concentrated around the average runtime.
* A small number of movies have exceptionally long runtimes, pulling the distribution toward the right.

This suggests that while the majority of films maintain a standard runtime, several lengthy productions exist within the dataset.

## Kurtosis Analysis

The runtime distribution demonstrates moderate tail behavior with the presence of several extreme observations.

Interpretation:

* Most observations are concentrated near the center.
* Some movies have unusually short or long runtimes.
* The distribution contains noticeable outliers, making it deviate from a perfectly normal distribution.

## Boxplot Analysis

The boxplot provides further evidence of outliers.

### Lower Outliers

Several movies have runtimes below 60 minutes, including one movie with a runtime close to 17 minutes.

These films may represent:

* Short films
* Documentary features
* Special releases
* Data recording anomalies

### Upper Outliers

Multiple movies exceed 200 minutes, with the maximum runtime reaching 321 minutes.

These may represent:

* Epic historical films
* Extended editions
* Multi-part productions
* Special theatrical releases

## Interquartile Range (IQR)

The middle 50% of movies lie between:

* Q1 = 118 minutes
* Q3 = 148 minutes

IQR:

```text
148 - 118 = 30 minutes
```

This indicates that the majority of movies have runtimes within a relatively narrow 30-minute interval, demonstrating consistency in film duration.

## Business Insight

The analysis reveals that the film industry tends to favor runtimes between 2 and 2.5 hours. Movies outside this range are relatively rare and appear as statistical outliers. The concentration around 130–140 minutes suggests an industry preference for feature-length films that balance storytelling depth with audience engagement.

## Conclusion

Runtime analysis indicates that movie durations are generally standardized around 133 minutes. Although the distribution is moderately right-skewed due to several exceptionally long films, most movies fall within a consistent runtime range of 118–148 minutes. The presence of both short-duration and long-duration outliers highlights the diversity of content available within the IMDb dataset.
